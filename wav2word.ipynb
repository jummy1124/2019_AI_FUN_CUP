```python
import speech_recognition

r=speech_recognition.Recognizer()

A=list()
B=list()
C=list()

for i in range(1,10):
    print(i)
    with speech_recognition.AudioFile("FUN CUP考試資料100題_A//A//A000000"+str(i)+".wav") as source_A:
        Topic_A=r.record(source_A)
        A.append(r.recognize_google(Topic_A,language='zh-tw'))

    with speech_recognition.AudioFile("FUN CUP考試資料100題_B_C_加密//B//B000000"+str(i)+".wav") as source_B:
        Question_B=r.record(source_B)
        B.append(r.recognize_google(Question_B,language='zh-tw'))

    with speech_recognition.AudioFile("FUN CUP考試資料100題_B_C_加密//C//C000000"+str(i)+".wav") as source_C:
        Choice_C=r.record(source_C)
        C.append(r.recognize_google(Choice_C,language='zh-tw'))    
        
        
for i in range(10,100):
    print(i)
    with speech_recognition.AudioFile("FUN CUP考試資料100題_A//A//A00000"+str(i)+".wav") as source_A:
        Topic_A=r.record(source_A)
        A.append(r.recognize_google(Topic_A,language='zh-tw'))

    with speech_recognition.AudioFile("FUN CUP考試資料100題_B_C_加密//B//B00000"+str(i)+".wav") as source_B:
        Question_B=r.record(source_B)
        B.append(r.recognize_google(Question_B,language='zh-tw'))

    with speech_recognition.AudioFile("FUN CUP考試資料100題_B_C_加密//C//C00000"+str(i)+".wav") as source_C:
        Choice_C=r.record(source_C)
        C.append(r.recognize_google(Choice_C,language='zh-tw'))
        
        
with speech_recognition.AudioFile("FUN CUP考試資料100題_A//A//A0000100.wav") as source_A:
    Topic_A=r.record(source_A)
    A.append(r.recognize_google(Topic_A,language='zh-tw'))

with speech_recognition.AudioFile("FUN CUP考試資料100題_B_C_加密//B//B0000100.wav") as source_B:
    Question_B=r.record(source_B)
    B.append(r.recognize_google(Question_B,language='zh-tw'))

with speech_recognition.AudioFile("FUN CUP考試資料100題_B_C_加密//C//C0000100.wav") as source_C:
    Choice_C=r.record(source_C)
    C.append(r.recognize_google(Choice_C,language='zh-tw'))
```
